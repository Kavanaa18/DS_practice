```python
import pandas as pd

file_path = '/content/drive/My Drive/datasets/Alcohol_Sales.csv'


df = pd.read_csv(file_path , index_col = "DATE" , parse_dates = True)
df.head()
```

```python
df.columns = ["Sales"]
df.plot(figsize=(12,5))
```

```python
df
```

```python
#creating 3 columns based on the previous data

df["Sales_lastmonth"] = df["Sales"].shift(+1)
df["Sales_2Monthsback"]  =df["Sales"].shift(+2)
df["Sales_3Monthsback"]  =df["Sales"].shift(+3)
df
```

```python
df= df.dropna()
```

```python
df
```

```python
# linear regression

from sklearn.linear_model import LinearRegression

lin_model = LinearRegression()
```

```python
#RandomForest regressor

from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=100 , max_features=3 , random_state=1)
```

```python
#splitting the data
import numpy as np

x1,x2,x3,y=df["Sales_lastmonth"] , df["Sales_2Monthsback"],df['Sales_3Monthsback'],df['Sales']
x1,x2,x3,y = np.array(x1) , np.array(x2),np.array(x3),np.array(y)
x1,x2,x3,y = x1.reshape(-1,1),x2.reshape(-1,1),x3.reshape(-1,1),y.reshape(-1,1)
final_x = np.concatenate((x1,x2,x3),axis=1)

print(final_x)
```

```python
x_train,x_test,y_train,y_test = final_x[:-30] , final_x[-30:],y[:-30],y[-30:]
```

```python
rf_model.fit(x_train,y_train)
lin_model.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
rf_pred = rf_model.predict(x_test)

plt.plot(rf_pred,label = "Randomforest regressor predictions")
plt.plot(y_test,label = "Actual values")
plt.legend(loc= "upper left")
plt.show()
```

```python
lin_pred = lin_model.predict(x_test)
plt.plot(lin_pred , label="Linear regression predictions")
plt.plot(y_test , label="Actual value")
plt.legend(loc="upper left")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

from math import sqrt

rmse_rf = sqrt(mean_squared_error(rf_pred,y_test))
rmse_lin = sqrt(mean_squared_error(lin_pred  , y_test))
```

```python
print("rmse_rf : ", rmse_rf)
print("rmse_lin : ", rmse_lin)
```
