```python
# filter operation

# salary of the people less than or equal to 20000

df.filter("total_bill >40").show()
```

```python
df.filter("total_bill >40").select(["total_bill","sex"]).show()
```

```python
df.filter((df["total_bill"]>40) & (df["size"]>3)).select(["total_bill" , "size"]).show()
```

```python
df.filter(df["sex"]=="Male").select(df["sex"]).show()
```

```python
df.filter((df["total_bill"]>40) | (df["size"]>3)).select(["total_bill" , "size"]).show()
```

```python
#not operator ~
df.filter(~(df["sex"]=="Male")).show()
```
